```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(x)
```

```python
plt.plot(x, '.')
```

```python
plt.hist(x)
```

```python
x = np.random.normal(scale=0.3, size=1000)
plt.plot(x, 'go')
```

```python
plt.hist(x)
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'g', linestyle='-.')
plt.plot(x, np.cos(x), color = [1, 0, 0], linestyle='--')

plt.show()
```

```python
plt.figure(figsize=(10, 8)) # image for several plots

plt.subplot(2, 2, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

plt.subplot(2, 2, 2)
plt.plot(x, np.exp(x), '--', color = 'r')

plt.subplot(2, 2, 3)
plt.plot(x, np.cos(x), '.', color = 'g')

plt.subplot(2, 2, 4)
plt.plot(x, np.abs(x), '-.', color = 'b')
```

```python
plt.plot(x, np.sin(x), color = 'g', linestyle='-.', label="sin(x)")
plt.plot(x, np.cos(x), color = [1, 0, 0], linestyle='--', label="cos(x)")

plt.legend()

plt.title("trigonometric functions")

plt.xlabel("x")
plt.ylabel("y")

plt.savefig('../data/images/tfs.png')

plt.show()
```

```python
x1 = np.random.normal(1, 1, 10000)
x2 = np.random.normal(-1, 1, 10000)

plt.hist(x1, color = 'g', alpha = 0.5, bins=100)
plt.hist(x2, color = 'r', alpha = 0.5, bins=100)

plt.show()
```

```python
def f(x, y):
    return np.sin(x) + np.cos(y)
```

```python
x = np.linspace(0, 2*np.pi, 1000)
X, Y = np.meshgrid(x, x)
```

```python
X
```

```python
Y
```

```python
Z = f(X, Y)
plt.style.use('seaborn-white')
plt.contourf(x, x, Z, cmap='RdBu_r')
```

# Seaborn

```python
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv('../data/titanic_train.csv')
df.head()
```

```python
sns.set(style="whitegrid", font_scale=1.5)
sns.set_color_codes("bright")
```

```python
sns.distplot(df['Fare'])
```

```python
ax = sns.distplot(df['Fare'])
ax.set(xscale='log')
```

```python
plt.figure(figsize=(6, 5))

sns.countplot(x='Embarked', data=df)
```

```python
plt.figure(figsize=(6, 5))

sns.barplot(x=df['Embarked'], y=df['Age'])

plt.title('       (+CI)')

plt.xlabel(' ')
plt.ylabel('')
```

```python
sns.boxplot(x=df['Age'], y=df['Sex'])
```

```python
sns.boxplot(x=df['Fare'], y=df['Pclass'], orient='h')
```

```python
sns.scatterplot(df['Age'], df['Fare'])
```

```python
sns.countplot(x=df['Pclass'], hue=df['Survived'])
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
df.columns
```

```python
sns.catplot(x='Pclass', y='Age', data=df)
```

```python
# kind : string, optional
#     The kind of plot to draw (corresponds to the name of a categorical
#     plotting function. Options are: "point", "bar", "strip", "swarm",
#     "box", "violin", or "boxen".
sns.catplot(x='Pclass', y='Age', data=df, kind='violin')
```
